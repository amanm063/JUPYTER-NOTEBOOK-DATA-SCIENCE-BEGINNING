```python
age = []
sex=[]
bmi = []
children = []
smoker = []
region = []
charges =[]
```

```python
def list_maker(lst,csv_file,clm):
    with open(csv_file) as csv_info:
        csv_dict = csv.DictReader(csv_info)
        for row in csv_dict:
            lst.append(row[clm])
        return lst
```

```python
list_maker(age, 'insurance.csv', 'age')
```
